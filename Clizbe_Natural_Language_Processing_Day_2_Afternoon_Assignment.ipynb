### Read the CNN Lite plain text file articles into a corpus using the NLTK's PlaintextCorpusReader.

```python
PATH = '/content/drive/My Drive/DSI Month 4/Week 13/NLP text sets/cnn_lite'

DOC_PATTERN = r'.*\.txt'
corpus = PlaintextCorpusReader(PATH, DOC_PATTERN)
```

### Iterate through the fileids in the corpus, extract the raw text of each document, and store them in a list.

```python
docs = [corpus.raw(fileid) for fileid in corpus.fileids()]
```

```python
docs[:5]
```

### Preprocess and clean the documents according to the steps below.

- Word Tokenize
- Lowercase
- Remove Stopwords
- Remove Punctuation
- Lemmatize
- Stem

```python
lemmatizer = WordNetLemmatizer()
stemmer = SnowballStemmer('english')

preprocessed = []

for doc in docs:
  tokenized = word_tokenize(doc)

  clean = [stemmer.stem(lemmatizer.lemmatize(token.lower())) 
            for token in tokenized 
            if token.lower() not in stopwords.words('english') 
            if token.isalpha()]
  # clean = []
  # for token in tokenized:
  #   if token.lower() not in stopwords.words('english') and token.isalpha()==True:
  #     if token.isalpha()==True:
  #       lemmatized = lemmatizer.lemmatize(token.lower())
  #       stemmed = stemmer.stem(lemmatized)
  #       clean.append(stemmed)

  untokenized = ' '.join(clean)
  preprocessed.append(untokenized)

```

### Count vectorize the preprocessed documents.

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
vectors = vectorizer.fit_transform(preprocessed)

count = pd.DataFrame(vectors.toarray(), 
                     columns=vectorizer.get_feature_names())
```

### One hot vectorize the preprocessed documents.

```python
vectorizer = CountVectorizer(binary=True)
vectors = vectorizer.fit_transform(preprocessed)

one_hot = pd.DataFrame(vectors.toarray(), 
                     columns=vectorizer.get_feature_names())
```

```python
one_hot.head(10)
```

### TF-IDF vectorize the preprocessed documents.

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
vectors = vectorizer.fit_transform(preprocessed)

tfidf = pd.DataFrame(vectors.toarray(), 
                     columns=vectorizer.get_feature_names())

tfidf.head(10)
```

### Use Doc2Vec to vectorize the preprocessed documents.

Set the size of the vectors to be the same size as those of the other methods using the `vector_size` argument.

```python
from gensim.models.doc2vec import Doc2Vec, TaggedDocument

documents = [TaggedDocument(doc, [i]) 
             for i, doc in enumerate(preprocessed)]
```

```python
model = Doc2Vec(documents, vector_size=4654)

doc2vec = pd.DataFrame([[document]+list(model[document]) 
                        for document in range(len(docs))]).drop(0, axis=1)

doc2vec.head(10)
```
