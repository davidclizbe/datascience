```
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
```

```
df_data = pd.read_csv('https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/parkinsons.data')
df_data.head()
```

```
df_data.drop(['name'], axis=1, inplace=True)
df_data.head()
```

```
y = df_data['status']

y.head()
```

```
X = df_data.drop(['status'], axis=1)
X.head()
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```
X_train.head()
```

```
y_train.head()
```

```
clf = LogisticRegression(solver= 'lbfgs', random_state=0).fit(X_train, y_train)

'lbfgs'
```

```
clf.score(X_train, y_train)
```

```
clf.score(X_test, y_test)
```
