```
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import seaborn as sn
```

```
data = pd.read_csv('https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/breast-cancer-wisconsin-data.csv')
data
```

```
X = data.drop(['id', 'Unnamed: 32', 'diagnosis'], axis=1)
X
```

```
y = data['diagnosis']
y
```

```
X_scaled = StandardScaler().fit_transform(X)

pca = PCA(n_components=2)
pcadata = pca.fit_transform(X_scaled)
```

```
sn.scatterplot(pcadata[:,0],pcadata[:,1] , hue = y)
```

```
from sklearn.manifold import TSNE

X_trans = TSNE(n_components=2).fit_transform(X_scaled)
```

```
sn.scatterplot(X_trans[:,0], X_trans[:,1] , hue = y)
```

```
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize = (16,8))
sn.scatterplot(pcadata[:,0],pcadata[:,1] , ax = axes[0], hue = y)
axes[0].set_title('PCA')
sn.scatterplot(X_trans[:,0], X_trans[:,1] , ax = axes[1], hue = y)
axes[1].set_title('TSNE')
```

```
X_trans = TSNE(n_components=2, perplexity= 5).fit_transform(X_scaled)
```

```
sn.scatterplot(X_trans[:,0], X_trans[:,1] , hue = y)
```

```
X_trans = TSNE(n_components=2, perplexity= 50).fit_transform(X_scaled)
```

```
sn.scatterplot(X_trans[:,0], X_trans[:,1] , hue = y)
```

```
X_trans = TSNE(n_components=2, perplexity= 50, n_iter = 2000).fit_transform(X_scaled)
```

```
sn.scatterplot(X_trans[:,0], X_trans[:,1] , hue = y)
```

```
X_trans = TSNE(n_components=2, perplexity= 50, n_iter = 250).fit_transform(X_scaled)
```

```
sn.scatterplot(X_trans[:,0], X_trans[:,1] , hue = y)
```
